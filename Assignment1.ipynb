```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data/fueltypesall.csv')
df['Date'] = pd.to_datetime(df.Date)
df
```

## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2019. Plot the corresponding bar chart.

```python
#Filtering Dates & FuelType
ts1 = pd.to_datetime('1/1/2019')
ts2 = pd.to_datetime('12/31/2019')

after_start = df["Date"] >= ts1
before_end = df["Date"] <= ts2
fuelType = (df['Fuel Type'] == 'Regular Unleaded Gasoline')
between = after_start & before_end & fuelType
diffCities = ["Ottawa","Toronto West/Ouest","Toronto East/Est","Windsor","London","Sudbury","Sault Saint Marie","Thunder Bay","North Bay","Timmins"]

filter = df.loc[between,:]

o = filter["Ottawa"].mean()
tw = filter["Toronto West/Ouest"].mean()
te = filter["Toronto East/Est"].mean()
w = filter["Windsor"].mean()
l = filter["London"].mean()
sb = filter["Sudbury"].mean()
ssm = filter["Sault Saint Marie"].mean()
tb = filter["Thunder Bay"].mean()
nb = filter["North Bay"].mean()
t = filter["Timmins"].mean()

graphData = pd.DataFrame({"Cities":["Ottawa","Toronto West/Ouest","Toronto East/Est","Windsor","London","Sudbury","Sault Saint Marie","Thunder Bay","North Bay","Timmins"],"Means":[o,tw,te,w,l,sb,ssm,tb,nb,t]})
finalGraph = graphData.plot.barh(x = 'Cities',y = 'Means',rot=5)

finalGraph.set_title("Means of the Regular Unleaded Gasoline prices")





```

## Question 3
Plot the boxplot for the Regular Unleaded Gasoline price in Ottawa across year 2012-2019. The distribution of each year should be visualized seperately.

```python


color = '#2C7BB6'
for i in range(8):
    bp = df[df["Fuel Type"] == "Regular Unleaded Gasoline"]
    bp = bp[pd.DatetimeIndex(bp["Date"]).year == 2012 + i]
    bp = bp["Ottawa"]
    plt.boxplot(bp,positions = [2012 + i], widths = 0.5)
    plt.title("Regular Unleaded Gasoline price in Ottawa across year 2012-2019")
    plt.xlabel("Year")
    plt.ylabel("Gas")
    
```
